```python
# filtering the columns that can be used futher for analysis.
# geners
# id
# keywords
# title
# overview
# cast
# crew
```

```python
df = movies_df[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
# use the literal_eval function. Originaly the data is in string format so convert it to list.
import ast
def convert(obj):
  L = []
  for i in ast.literal_eval(obj):
    L.append(i['name'])
  return L
```

```python
df['genres'] = df['genres'].apply(convert)
```

```python
df.head()
```

```python
df['keywords'] = df['keywords'].apply(convert)
```

```python
df.head()
```

```python
df.cast[0]
```

```python
#lets extract top 3 actor names from cast column
def extract(obj):
  L= []
  counter = 0
  for i in ast.literal_eval(obj):
    if counter != 3:
      L.append(i['name'])
      counter+=1
    else:
      break
  return L    
```

```python
df['cast'].apply(extract)
```

```python
df['cast'] = df['cast'].apply(extract)
```

```python
df['crew'][0]
```

```python
# fetching the name of the director
def fetch_director(obj):
  L =[]
  for i in ast.literal_eval(obj):
    if i['job'] == 'Director':
      L.append(i['name'])
      break
  return L 
```

```python
df['crew'].apply(fetch_director)
```

```python
df['crew'] = df['crew'].apply(fetch_director)
```

```python
# the overview is in string format
df['overview'][0]
```

```python
df['overview'] = df['overview'].apply(lambda x:x.split())
```

```python
# perform transformation and remove spaces
df['genres'] = df['genres'].apply(lambda x:[i.replace(" ","") for i in x])
df['keywords'] = df['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
df['cast'] = df['cast'].apply(lambda x:[i.replace(" ","") for i in x])
df['crew'] = df['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
df.head()
```

```python
df['tags'] = df['overview'] + df['genres'] + df['keywords'] + df['cast'] + df['crew']
```

```python
df.head()
```

```python
# dropping columns like overview,cast,crew,genres,keywords
new_df = df[['movie_id','title','tags']]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
new_df.head()
```

```python
# converting the string to lower case
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
# perform stemming 
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  y = []
  for i in text.split():
    y.append(ps.stem(i))

  return " ".join(y)  
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
# most of the data in the dataframe is in text format.Here we need to vectorize this first and then find the similarity among the movies. 
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words ='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
# finding the cosine distance among the 5000 Dimension
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
similarity[0]

#this array of first movie denotes its similarity with rest of 4809 movies 
```

```python
# define function to recommend movies
def recommend(movies):
  movie_index = new_df[new_df['title']== movies].index[0]
  distances = similarity[movie_index]
  movie_list = sorted(list(enumerate(distances)),reverse = True,key=lambda x:x[1])[1:6]


  for i in movie_list:
    print(new_df.iloc[i[0]].title)
    

```

```python
recommend('Batman Begins')
```
